## 1. 線形単回帰分析クラスの実装
***

$$
y = ax + b
$$

を目的の直線とする線形回帰分析について考える．

この関数における「傾き$a$」と「切片$b$」をデータセット内のデータから求めるとき，最小化したい値は，

$$
E = \sum_{i=1}^n(ax_i + b - y_i)^2
$$

となり，これは「実際の目的変数」と「予測された目的変数」の**残差平方和**である．

$x_i$と$y_i$の組はデータセットから得られる値であるため，$E$は$a$と$b$の関数であると言える．この$E$を最小化するためには，$a$と$b$それぞれに対する偏微分を行い，それらが$0$である時を求めることとなるので，

$$
\frac{\partial E}{\partial a} = 0
$$

$$
\frac{\partial E}{\partial b} = 0
$$

となる．つまり，

$$
\frac{\partial E}{\partial a} = \sum_{i=1}^n 2x_i(ax_i + b - y_i) = 0
$$

$$
\frac{\partial E}{\partial b} = \sum_{i=1}^n (ax_i + b - y_i) = 0
$$

であり，これを$b$について整理すると，

$$
b = \frac{1}{n}\sum_{i=1}^n (y_i - ax_i) \quad{(1)}
$$

となる．また，$a$に関しては，

$$
a = \frac{\sum_{i=1}^n x_iy_i - \frac{1}{n} \sum_{i=1}^n x_i \sum_{i=1}^n y_i} {\sum_{i=1}^n x_i^2 - \frac{1}{n}(\sum_{i=1}^n x_i)^2} \quad{(2)}
$$

つまり，式(1), (2)を用いることで，データセット内のデータから線形回帰直線を求めることができる．

また，決定係数に関しては，以下の式で求めることができる．

$$
R^2 = 1 - \frac{\sum_{i=1}^n (y_i - \hat{y_i})^2}{\sum_{j=1}^n (y_j - \bar{y})^2} \quad{(3)}
$$

ここで，$y_i$は「実際の目的変数の値」，$\hat{y_i}$は「予測された目的変数の値」，$\bar{y}$は「実際の目的変数の平均値」である．
