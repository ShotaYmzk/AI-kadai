データフレームは以下の通りとします。

||ワンルーム|1k|1LDK|2LDK|3LDK|
|:---:|:---:|:---:|:---:|:---:|:---:|
|港区|7.5|9.6|17.5|24.5|31.9|
|世田谷区|7.0|8.0|11.6|14.9|18.3|
|台東区|7.8|8.7|12.4|15.6|19.0|
|新宿区|7.5|8.9|14.3|19.1|24.2|
|渋谷区|7.8|9.5|16.2|22.1|28.4|
|目黒区|7.7|8.9|13.5|17.5|21.8|

```python
"""
@@@ ヒント @@@

データフレームとは、縦横の長さが一定の表データのことです。
縦列を column、横列を index と呼びます。

縦列と横列のキーがない表データ data がある場合、pd.DataFrame(dic, columns=col, index=idx) のようにするとデータフレームが作成できます。
- data ... 縦列と横列のキーがない表データ (2次元配列)
- col  ... 縦列 (column) のキーのリスト
- idx  ... 横列 (index) のキーのリスト

このときのデータフレームの構造は以下のようになります。
       | col[0]       col[1]     ...
-------+-----------------------------
idx[0] | data[0][0]   data[0][1] ...
idx[1] | data[1][0]   data[1][1] ...
...

"""

# 東京の家賃相場レコード
dic = {
    "ワンルーム":[7.5, 7.0, 7.8, 7.5, 7.8, 7.7],
    "1k":[9.6, 8.0, 8.7, 8.9, 9.5, 8.9],
    "1LDK":[17.5, 11.6, 12.4, 14.3, 16.2, 13.5],
    "2LDK":[24.5, 14.9, 15.6, 19.1, 22.1, 17.5],
    "3LDK":[31.9, 18.3, 19.0, 24.2, 28.4, 21.8]
}
idx = ["港区", "世田谷区", "台東区", "新宿区", "渋谷区", "目黒区"]
df_tky:pd.DataFrame = pd.DataFrame(dic, index=idx)
print(df_tky)
```
